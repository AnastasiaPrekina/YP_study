### last_price

Цена на момент снятия с публикации. Тип данных - float, напрашивается поменять, но на самом деле для упрощенного восприятия я предалагаю оставить его с этим типом, и поделить все значения на 1 000 000. В задании не написано, но все-таки будем считать, что цены были в рублях, а ценами за недвижимость легче оперировать в миллионах и тысячах, да и редко кто указывает цену с прицнипиальностью до сотен рублей. Поэтому дополнительно округлим до 3х знаков после запятой.

```python
data['last_price'] = (data['last_price'] / 1000000).round(3)
data
```

```python
data['last_price'].describe().round(3)
```

Видим нечто странное, некий объект недвижимости стоил 12 тысяч рублей, цена аренды(?), к максимальному тоже можно придраться, но для объективности посмотрим на распределение значений на графике.

```python
data.boxplot(column='last_price', figsize=(9, 9))
```

Видим, что на графике есть выбросы, особенно выделяется как раз макисмальное значение. Предлагаю разбить данные по стоимости на несколько категорий, и, скорее всего избавиться от аномалий.

```python
data['last_price'].hist(range=(0, 1), bins=100) #смотрим распределение цен до 1млн
```

```python
data['last_price'].hist(range=(1, 10), bins=100) #смотрим распределение цен до 10млн
```

```python
data['last_price'].hist(range=(10, 50), bins=100) #смотрим распределение цен от 10 до 50млн
```

```python
data['last_price'].hist(range=(50, 800), bins=100) #смотрим распределение цен свыше 50 млн
```

Итак наше распределение похоже на нормальное, с очень длинным хвостом справа. Также видим, что значение в 12тыс.руб это выброс и скорее всего просто ошибочное. Значение в 763млн может быть настоящим, но оно единичное.  Предалагаю для более объективного исследования избавиться от хвостов и оставить строки с ценой в интервале от 400тыс до 50млн.
```python
#отбираем строки по условию, суммируем их количество и смотрим %  в общем количестве строк
data.query(('last_price<0.4') and ('last_price>50'))['last_price'].count() / data['last_price'].count() * 100
```

Тогда обозначим как наиболее большое реальное значение высоты потолка 7м. Нет, наверняка, в жизни встречается всякое, но это такие единичные случаи, что лучше нам на них не ориентироваться.

```python
data.pivot_table(index='locality_name', values='city_centers_nearest', aggfunc='median').head(10)
```

```python
# получим список городов, в которых есть пропуски в расстояниях
locality_list = []
for i in data['locality_name'].unique():
    if data[data['locality_name'] == i]['city_centers_nearest'].isna().sum() > 0:
        locality_list.append(i)
```

Не стала выводить списки с количеством и перечислением этих городов тк их очень много и это целая простыня.

```python
median_data = data.groupby('locality_name')['city_centers_nearest'].median() #определение медианного расстояния
for city in locality_list:                       #замена пустых ячеек медианным расстоянием по данной категории кроме СПб
    if city == 'Санкт-Петербург':
        data.loc[data['locality_name'] == city, 'city_centers_nearest'] = \
        data.loc[data['locality_name'] == city, 'city_centers_nearest'].fillna(-1000)
    else:
        median_nearest = median_data.loc[city]
        data.loc[data['locality_name'] == city, 'city_centers_nearest'] = \
        data.loc[data['locality_name'] == city, 'city_centers_nearest'].fillna(median_nearest)
```

```python
data['city_centers_nearest'].isna().sum()
```

```python
def floor_type(floor):          #функция для определения типа этажа
    if floor['floor'] == 1:
        return 'первый'
    if 1 < floor['floor'] < floor['floors_total']:
        return 'другой'
    if floor['floor'] == floor['floors_total']:
        return 'последний'
```

```python
data['floor_type'] = data.apply(floor_type, axis=1)
data['floor_type'].isna().sum()
```

В ходе исследования были рассмотрены факторы, которые имеют наибольшее влияние на формирование цены:
* площадь
* число комнат
* высота потолков
* удаленность от центра
* этаж

Общая цена коррелирует достаточно выраженно с общей площадью, при этом цена за метр квдратный уже слабо зависит от нее.
Цена имеет отрицательную корреляцию с удаленностью от центра, также как и цена за м2. Чем больше расстояние от центра, тем дешевле недвижимость.
Общая цена зависит от количества комнат, а вот цена за м2 практически не зависит от него.
Также можно отметить общий факт для всех факторов, чем больше общая площадь квартиры или ее общая стоимость тем непредсказуемее взаимосвязи всех факторов.

Был составлен топ-лист по количеству объявлений по Спб и ЛО и средней цене м2 в них. Помимо Спб большая часть квартир приходится на самые ближайшие районы-пригороды, такие как Мурино и Кудрово. Самые дорогие метры в Питере и Пушкине, а самые дешевые в Выборге.

Была определена центральная зона города в радиусе 7 км от центра города. Недвижимость в данном районе имеет схожие зависимости с общими результатами, но и имеет свои особенности. Так, например, она оказалась меньше подвержена спаду цен в 2016-2017 годах, цена за м2 падает при количестве комнат 2-6, расстояние до центра уже внутри центра является менее значимым фактором. Также квартиры на первых и последних этажах оцениваются в целом дешевле, чем на средних.

Для более точно определения зависимостей и предсказания цены на недвижимость было бы необходимо рассматривать совокупности и взаимосвязи всех факторов и их общее итоговое влияние.
